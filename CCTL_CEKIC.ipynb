### Fonctionnement de base de l'algorithme du Perceptron
Le perceptron fonctionne en recevant des entrées (features), chacune associée à un poids, et en les combinant dans une somme pondérée. Cette somme est ensuite passée à travers une fonction d'activation (généralement une fonction d'activation de type seuil) pour produire une sortie. La sortie du perceptron est souvent utilisée pour prédire si une instance de données appartient à l'une des deux classes possibles (par exemple, 0 ou 1)sage)..

### Comment il apprend à classer les données
L'apprentissage dans un perceptron simple couche se fait par un processus itératif où l'algorithme tente de minimiser les erreurs de classification. Lors de l'entraînement :
1. **Initialisation** : Les poids sont initialement définis à de petites valeurs aléatoires. 
2. **Pour chaque instance dans l'ensemble d'entraînement** : Le perceptron calcule la sortie en fonction des entrées actuelles et des poids actuels. 
3. **Calcul de l'erreur** : La sortie calculée est comparée à la sortie attendue (vraie valeur), et l'erreur est calculée. 
4. **Mise à jour des poids** : Les poids sont ajustés pour réduire l'erreur. Cela se fait en ajoutant à chaque poids une fraction de l'entrée correspondante multipliée par l'erreur et par le taux d'apprentissage (un petit nombre positif qui contrôle la vitesse d'apprentissage).

L'intuition derrière cette règle est simple : si la prédiction est correcte, le changement dans les poids sera faible. Cependant, si la prédiction est incorrecte, les poids seront ajustés de manière significative dans la direction qui réduit l'erreur de prédiction. Ce processus se répète pour de nombreuses itérations ou jusqu'à ce qu'un critère d'arrêt spécifique soit atteint (par exemple, un certain nombre d'itérations ou un seuil d'erreur minimale).

## **Partie 3 : Calculer la régression logistique sur un jeu de données**

- Utiliser le jeu de données fourni pour effectuer une classification binaire (intérêt pour le service haut débit).
- Diviser les données en ensembles d'entraînement et de test.
- Entrainer le modèle de régression logistique sur l'ensemble d'entraînement.
- Evaluer les performances du modèle sur l'ensemble de test en utilisant des métriques telles que la précision, le rappel et la courbe ROC.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score, roc_curve
import matplotlib.pyplot as plt
import warnings

# Ignorer les avertissements
warnings.filterwarnings('ignore')

# Création d'un DataFrame avec les données
data = {
    'Age': [30, 25, 40, 35, 28, 45, 22, 38, 32, 29],
    'Revenu_mensuel_euros': [2500, 1800, 3500, 2900, 2000, 4000, 1500, 3200, 2800, 2200],
    'Duree_abonnement_mois': [12, 6, 24, 18, 10, 36, 3, 20, 15, 8],
    'Score_satisfaction': [8, 6, 9, 7, 5, 10, 4, 8, 6, 7],
    'Interet_service': [1, 0, 1, 1, 0, 1, 0, 1, 0, 0]
}
df = pd.DataFrame(data)

# Séparation des données
X = df[['Age', 'Revenu_mensuel_euros', 'Duree_abonnement_mois', 'Score_satisfaction']]
y = df['Interet_service']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Entraînement du modèle de régression logistique
model = LogisticRegression()
model.fit(X_train, y_train)

# Prédictions et calcul des scores
y_pred = model.predict(X_test)
y_pred_proba = model.predict_proba(X_test)[:, 1]
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_proba)

# Affichage des scores
print(f"Exactitude : {accuracy:.2f}")
print(f"Précision : {precision:.2f}")
print(f"Rappel : {recall:.2f}")
print(f"AUC-ROC : {roc_auc:.2f}")

# Génération de la courbe ROC
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
plt.figure()
plt.plot(fpr, tpr, label='Régression logistique (AUC = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Taux de faux positifs')
plt.ylabel('Taux de vrais positifs')
plt.title('Caractéristique de fonctionnement du récepteur (ROC)')
plt.legend(loc="lower right")
plt.show()
```
